```python
%pylab inline
from astropy.io import fits
```

```python
from pyraf import iraf
iraf.kepler(_doprint=0)
```

```python
shortname, period = 'j900', 6.871311

iraf.unlearn('kepfold')
with matplotlib.rc_context():
    iraf.kepfold(
        infile='./{0}/{0}_sff.fits'.format(shortname),
        outfile='./{0}/{0}_folded_{1:2.3f}.fits'.format(shortname, period),
        period=period,
        bindata=True,
        binmethod="median",
        threshold=2.0,
        niter=5,
        nbins=100,
        quality=True,
        plottype='det',
        clobber=True,
    )
```

```python
j900 = fits.open('./j900/j900_folded_6.871.fits')
j900_folded = j900[1].data
```

```python
plt.scatter(j900_folded['PHASE'], j900_folded['DETSAP_FLUX'], alpha=0.2)
plt.scatter(j900_folded['PHASE'] + 1.0, j900_folded['DETSAP_FLUX'], alpha=0.2)
plt.title('J900 folded at {:2.3f} days'.format(period))
plt.xlim(0, 1.5)
plt.ylabel('Normalized Flux')
plt.xlabel('Phase (radians)')
plt.xticks([0,0.25,0.5,0.75,1.0,1.25,1.5], ['0',  '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$', '$5\pi/2$', '$3\pi$'])
```
