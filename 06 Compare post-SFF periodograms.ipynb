```python
%pylab inline
from astropy.io import fits
from astropy.table import Table
```

```python
# Write out SFFed SAP_FLUX for good quality points
for shortname in ('k371', 'j900', 'hocr1'):
    data = fits.open('./{0}/{0}_sff.fits'.format(shortname))[1].data
    times = data['TIME'][data['SAP_QUALITY'] == 0]
    sap_fluxes = data['SAP_FLUX'][data['SAP_QUALITY'] == 0]
    np.savetxt('{0}/{0}_sff_sap.csv'.format(shortname), zip(times, sap_fluxes))
```

```python
# Write out SFFed DETSAP_FLUX for good quality points
for shortname in ('k371', 'j900', 'hocr1'):
    data = fits.open('./{0}/{0}_sff.fits'.format(shortname))[1].data
    times = data['TIME'][data['SAP_QUALITY'] == 0]
    sap_fluxes = data['DETSAP_FLUX'][data['SAP_QUALITY'] == 0]
    np.savetxt('{0}/{0}_sff_detsap.csv'.format(shortname), zip(times, sap_fluxes))
```

# After running those through the [periodogram service](http://exoplanetarchive.ipac.caltech.edu/cgi-bin/Periodogram/nph-simpleupload), have a look at the results

The service produces `.csv.out` files for each `.csv`

```python
pgrams = {}
for shortname in ('k371', 'j900', 'hocr1'):
    these_pgrams = pgrams[shortname] = {}
    these_pgrams['original'] = Table.read('./{0}/{0}_original_meanflux.csv.out'.format(shortname), format='ipac')
    these_pgrams['sap'] = Table.read('./{0}/{0}_truncated.csv.out'.format(shortname), format='ipac')
    these_pgrams['sff_sap'] = Table.read('./{0}/{0}_sff_sap.csv.out'.format(shortname), format='ipac')
    these_pgrams['sff_detsap'] = Table.read('./{0}/{0}_sff_detsap.csv.out'.format(shortname), format='ipac')
```

```python
for shortname in ('k371', 'j900', 'hocr1'):
    p = pgrams[shortname]
    plt.figure(figsize=(14,5))
    
    plt.subplot(121)
    plt.title('Lomb-Scargle periodogram for {}'.format(shortname))
    plt.plot(p['original']['PERIOD'], p['original']['POWER'], label='Mean flux, whole frame')
    plt.plot(p['sap']['PERIOD'], p['sap']['POWER'], label='Simple aperture (SAP) fluxes')
    plt.plot(p['sff_sap']['PERIOD'], p['sff_sap']['POWER'], label='SAP + SFF')
    plt.plot(p['sff_detsap']['PERIOD'], p['sff_detsap']['POWER'], label='SAP + flattened + SFF')
    
    plt.legend(loc='upper left')
    plt.xscale('log')
    plt.xlabel('Period (days)')
    plt.ylabel('Power')
    plt.xlim(0.1, np.max(p['original']['PERIOD']))
    
    plt.subplot(122)
    data = fits.getdata('./{0}/{0}_sff.fits'.format(shortname), 1)
    sapsff = plt.plot(data['TIME'][data['SAP_QUALITY'] == 0], data['SAP_FLUX'][data['SAP_QUALITY'] == 0], color='blue', label="SAP + SFF")
    plt.legend()
    ax = plt.gca()
    ax2 = ax.twinx()
    sapflatsff = ax2.plot(data['TIME'][data['SAP_QUALITY'] == 0], data['DETSAP_FLUX'][data['SAP_QUALITY'] == 0], color='red', label="SAP + flattened + SFF")
    # hack to make legend show
    ax2.plot(data['TIME'][data['SAP_QUALITY'] == 0][0], data['DETSAP_FLUX'][data['SAP_QUALITY'] == 0][0], color='blue', label="SAP + SFF")
    plt.legend()
```

```python
p = pgrams['j900']
plt.figure(figsize=(7,5))

plt.title('Lomb-Scargle periodogram for J900')
plt.plot(p['sff_detsap']['PERIOD'], p['sff_detsap']['POWER'])

plt.xscale('log')
plt.xlabel('Period (days)')
plt.ylabel('Power')
plt.xlim(0.1, np.max(p['original']['PERIOD']))
plt.axvline(x=6.871311, color='red')
plt.annotate('T = 6.871 days', (6.87131 + 0.6, 95))
plt.savefig('j900/j900_sff_detsap_periodogram.pdf', bbox_inches='tight')
```
