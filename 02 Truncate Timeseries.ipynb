# Turn this shaded region in the plot into a time range to clip off

`kepclip` wants barycentric Julian dates, BJD. The `TIME` column is in BJD - BJDREFI + BJDREFF.

Get BJDREFI and add it to the xmin of that shaded range:

```python
bjd_excluded_end = 1941 + k371[1].header['BJDREFI'] + k371[1].header['BJDREFF']
print bjd_excluded_end
```

```python
from pyraf import iraf
iraf.kepler(_doprint=0)
```

```python
for shortname in ('k371', 'j900', 'hocr1'):
    iraf.unlearn('kepclip')
    iraf.kepclip(
        infile='{0}/{0}_sap.fits'.format(shortname),
        outfile='{0}/{0}_truncated.fits'.format(shortname),
        ranges='0,{}'.format(bjd_excluded_end),
        plot=False
    )
```
