# Flatten K 3-71

There is perhaps something to be gained from individually examining the timeseries as we flatten them instead of applying the same order polynomial and outlier rejection criteria across all 3 data sets.

```python
iraf.unlearn('kepflatten')
iraf.kepflatten(
    infile='k371/k371_truncated.fits',
    outfile='k371/k371_flattened.fits',
    datacol='SAP_FLUX',
    errcol='SAP_FLUX_ERR',
    nsig=2.0,
    stepsize=0.2,
    winsize=5.0,
    npoly=2,
    niter=10,
    ranges='0,0',
    plot=True
)
```

# Flatten J900

Increased winsize from 5.0 days to 7.0 days because it looked like this was overfitting

```python
iraf.unlearn('kepflatten')
iraf.kepflatten(
    infile='j900/j900_truncated.fits',
    outfile='j900/j900_flattened.fits',
    datacol='SAP_FLUX',
    errcol='SAP_FLUX_ERR',
    nsig=2.0,
    stepsize=0.2,
    winsize=7.0,
    npoly=2,
    niter=10,
    ranges='0,0',
    plot=True
)
```
