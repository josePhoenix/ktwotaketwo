# Compare Pre-SFF periodograms

One way to guard against introducing periodic variation as a side effect of data reduction is to have a "before" and "after" periodogram. It also helps to see the power spectrum across all three targets, since peaks present in all 3 are surely systematic and not real.

We have the overall mean flux in the postage stamp, as well as the simple aperture photometry flux to look at so far. Let's dump those to CSV and run them through the periodogram service.

```python
%pylab inline
from astropy.io import fits
from astropy.table import Table
```

```python
# Write out nanmean flux across whole image for each timestamp, only good quality points

for shortname in ('k371', 'j900', 'hocr1'):
    data = fits.open('./{0}/{0}_original.fits'.format(shortname))[1].data
    times = data['TIME'][data['QUALITY'] == 0]
    fluxes = data['FLUX'][data['QUALITY'] == 0]
    mean_fluxes = np.nanmean(fluxes, axis=(-1, -2))
    np.savetxt('{0}/{0}_original_meanflux.csv'.format(shortname), zip(times, mean_fluxes))
```

```python
# Write out SAP_FLUX for good quality points
for shortname in ('k371', 'j900', 'hocr1'):
    data = fits.open('./{0}/{0}_truncated.fits'.format(shortname))[1].data
    times = data['TIME'][data['SAP_QUALITY'] == 0]
    sap_fluxes = data['SAP_FLUX'][data['SAP_QUALITY'] == 0]
    np.savetxt('{0}/{0}_truncated.csv'.format(shortname), zip(times, sap_fluxes))
```

# Run through the [periodogram service](http://exoplanetarchive.ipac.caltech.edu/cgi-bin/Periodogram/nph-simpleupload)

This produces `.csv.out` files for the `.csv` files.

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(14,4))
for shortname in ('k371', 'j900', 'hocr1'):
    table_orig = Table.read('./{0}/{0}_original_meanflux.csv.out'.format(shortname), format='ipac')
    table_sap = Table.read('./{0}/{0}_truncated.csv.out'.format(shortname), format='ipac')
    ax1.plot(table_orig['PERIOD'], table_orig['POWER'], label=shortname)
    ax2.plot(table_sap['PERIOD'], table_sap['POWER'], label=shortname)
for ax in (ax1, ax2):
    ax.legend(loc='upper center')
    ax.set_xscale('log')
    ax.set_xlabel('Period (days)')
    ax.set_ylabel('Power')
ax1.set_xlim(0.1, np.max(table_orig['PERIOD']))
ax2.set_xlim(0.1, np.max(table_sap['PERIOD']))
ax1.set_title('Lomb-Scargle Periodogram of mean flux over image at each time')
ax2.set_title('Lomb-Scargle Periodogram of simple aperture flux values')
plt.tight_layout()
```
